# Filtration samples with selected diagnoses from annotaion table

```python
blood_all_annotation = pd.read_csv("../data/annotations_all.csv", index_col="Sample")
```

```python
datasets = [
    "GSE110685",
    "GSE112057",
    "GSE112087",
    "GSE117769",
    "GSE120178",
    "GSE122438",
    "GSE122459",
    "GSE123658",
    "GSE141529",
    "GSE147339",
    "GSE159225",
    "GSE161199",
    "GSE168522",
    "GSE168813",
    "GSE176260",
    "GSE177044",
    "GSE181228",
    "GSE183701",
    "GSE186507",
    "GSE208581",
    "GSE216678",
    "GSE231686",
    "GSE231691",
    "GSE235357",
    "GSE265975",
    "GSE270454",
    "GSE72509",
    "GSE79970",
    "GSE81259",
    "GSE90081",
    "PRJNA393498",
    "PRJNA495603",
]

ann = blood_all_annotation[blood_all_annotation["Dataset"].isin(datasets)]
ann.Dataset.unique()
```

```python
selected_diagnoses = [
    "Healthy",
    "Type_1_diabetes_mellitus",
    "Systemic_lupus_erythematosus",
    "Ulcerative_colitis",
    "Crohn_disease",
    "Rheumatoid_arthritis",
    "Kidney_failure",
    "Juvenile_idiopathic_arthritis",
    "Systemic_inflammatory_response_syndrome",
    "Systemic_sclerosis, Interstitial_lung_disease",
    "Dysthyroid_exophthalmos",
    "Primary_sclerosing_cholangitis, Ulcerative_colitis",
    "Amyotrophic_lateral_sclerosis",
    "Allergy",
    "Ankylosing_spondylitis",
    "Systemic_lupus_erythematosus, Pulmonary_fibrosis",
    "Multiple_sclerosis",
    "Systemic_sclerosis",
    "Alzheimer_disease",
    "Relapsing-remitting_multiple_sclerosis",
    "Idiopathic_pulmonary_fibrosis",
    "Immune_thrombocytopenia",
    "Discoid_lupus_erythematosus",
    "Primary_sclerosing_cholangitis",
    "Psoriasis",
    "Inflammatory_bowel_disease",
    "Subacute_cutaneous_lupus_erythematosus",
    "Immune_thrombocytopenic_purpura",
    "CINCA_syndrome",
    "VEXAS_syndrome",
    "Secondary_progressive_multiple_sclerosis",
    "Chronic_multifocal_osteomyelitis",
    "Multisystem_inflammatory_syndrome",
    "Arthropathic_psoriasis",
    "Generalized_pustular_psoriasis",
]

filtered_ann = ann[
    (ann["Diagnosis"].isin(selected_diagnoses))
    & (ann["Tissue_stimulation"] == "Without_stimulation")
    & (ann["Readcounts"] > 500000)
    & (~ann["Library_preparation_kit"].str.contains("Ion Torrent", na=False))
    & (ann["Sample_type"] == "Whole_blood")
]

print("Filtered annotations:")
print(filtered_ann["Diagnosis"].value_counts())
```

```python
filtered_ann.to_csv("../data/annotations_filtred.csv", index=True)
```

```python
all_samples = list(filtered_ann.index)
```
